```python
# function to find the length of longest increasing subarray
def lenOfLongIncSubArr(arr) :
    
    # if array is empty then longest increasing subarray will be of 0 length
    if arr == []:
        return 0
 
    global_max_len = 1          # global_max_len stores length of longest increasing subarray of all time
    cur_max_len = 1             # cur_max_len stores length of longest increasing subarray currently
      
    # traverse the array from the 2nd element
    for i in range(1, len(arr)) :
 
        # if current element if greater than previous
        # element, then this element helps in building
        # up the previous increasing subarray encountered
        # so far
        if (arr[i] > arr[i-1]) :
            cur_max_len = cur_max_len + 1
            
        else :
            # check if 'max' length is less than the length
            # of the current increasing subarray. If true, 
            # then update 'max'
            if (global_max_len < cur_max_len)  :
                global_max_len = cur_max_len 
 
            # reset 'len' to 1 as from this element
            # again the length of the new increasing
            # subarray is being calculated    
            cur_max_len = 1
         
         
    # comparing the length of the last
    # increasing subarray with 'max'
    if (global_max_len < cur_max_len) :
        global_max_len = cur_max_len
      
    # required maximum length
    return global_max_len
 

# Driver program to test above

if __name__=='__main__':
    
    arr = [2,2,2,2,2]

    print("Length of longest increasing subarray = ", lenOfLongIncSubArr(arr))
 
```

```python
# function to form the largest number possible from a list of non-negative integers

def largestNumber(array):
    
    #If there is only one element in the list, the element itself is the largest element.
    #Below if condition checks the same.    
    if len(array)==1:  
        return str(array[0])
    
    # Converting integers in the list to strings 
    for i in range(len(array)):
        array[i]=str(array[i])
    # [10,2]=>['10','2']
    #Second, we find the largest element by swapping technique.
    for i in range(len(array)):
        for j in range(i+1,len(array)):
            if array[j]+array[i]>array[i]+array[j]:
                array[i],array[j]=array[j],array[i]
    #['2', '10']
    #Refer JOIN function in Python
    result=''.join(array)
    #Edge Case: If all elements are 0, answer must be 0 
    if(result=='0'*len(result)):
        return '0'
    else:
        return result
        
        
if __name__ == "__main__":
    a = [3,30,34,5,9]
    print("Largest possible number = ", largestNumber(a))
```

```python
# import json library to parse json
import json

# import pandas library to write data in a csv file
import pandas as pd

# Open the JSON file and read its contents
with open('sample_json.json') as f:
    json_data = json.load(f)

# Now json_data contains the parsed JSON data

# Recover the list of servlet's from json_data
servlet_list = json_data['web-app']['servlet']

# Defining an empty list to store servlet-name and servlet-class
servlet_details = []     

# Iterate through the servlet_list and save each servlet-name and servlet-class
for s in servlet_list:
    servlet_details.append([s['servlet-name'],s['servlet-class']])
    
# Now servlet_details is a nested list containing all servlet-name and servlet-class pairs

# Converting the servlet_details nested list to pandas dataframe
servlet_details = pd.DataFrame(servlet_details, columns=['servlet-name','servlet-class'])

# Writing the servlet_details dataframe to a csv file
servlet_details.to_csv('servlet_details.csv',index=False)
```
